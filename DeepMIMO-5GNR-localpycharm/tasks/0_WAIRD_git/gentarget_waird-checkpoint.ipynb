```python
#依次按照每一个bsidx和ueidx读取出相应特征并最终生成csv文件
for idx_filename, file_name in enumerate(tqdm(sorted_image_files)):
    # Extract numbers using regular expressions
    match = re.search(r'(\d+)_ue(\d+)\.png', file_name)
    if match:
        #打印出bsidx和ueidx
        bsIdx = int(match.group(1))
        ueIdx = int(match.group(2))
        # print(ueIdx)
        curEnv = 0
        envImg = envs[curEnv]
        linklocA = linklocAs[curEnv]
        ImgSize = linklocA[:2].numpy().astype(int)
        if scenario == 1:
            linklocA = torch.reshape(linklocA[2:], (150, 4))
            BSloc = linklocA[::30, :2][BSlist[bsIdx], :]
            UEloc = linklocA[UElist[ueIdx], 2:]
        else:
            BSloc = linklocA[2:4]
            linklocA = torch.reshape(linklocA[4:], (10000, 2))
            UEloc = linklocA[UElist[ueIdx], :]
        distance = distances[curEnv][0:1, BSlist[bsIdx] * 30 + UElist[ueIdx]]
        gain = gains[curEnv][0:1, BSlist[bsIdx] * 30 + UElist[ueIdx]] / (-200)
        angle = angles[curEnv][0, BSlist[bsIdx] * 30 + UElist[ueIdx]] / (-5)
        channel = readChannel(readImage, generatedFolder, BSlist[bsIdx], UElist[ueIdx], scenario)
        channel = channel / np.max(np.abs(channel))
        channel = torch.FloatTensor(
            np.concatenate((np.real(channel)[None, :, :], np.imag(channel)[None, :, :]), axis=0))
        sight = sights[curEnv][0:1, BSlist[bsIdx] * 30 + UElist[ueIdx]]
        scale = torch.FloatTensor([imageSize / np.max(ImgSize)])
        df.loc[len(df.index)] = [file_name, sight.numpy(), scale.numpy(), UEloc[0].numpy(), UEloc[1].numpy(), angle[0].numpy(), angle[1].numpy(), distance.numpy(), gain.numpy()]
#保存成csv文件
df.to_csv(targetfile, index=False)
        #添加到dataframe结构
        # return envImg, distance, channel, BSloc, UEloc, sight, scale, angle, gain
df
```
